# Data Preparation

```python
from bs4 import BeautifulSoup
import pandas as pd

import requests
```

```python
# Data source comes from Myhora

URL = "https://www.myhora.com/%E0%B8%AB%E0%B8%A7%E0%B8%A2/%E0%B8%AA%E0%B8%96%E0%B8%B4%E0%B8%95%E0%B8%B4%E0%B8%AB%E0%B8%A7%E0%B8%A2-%E0%B8%A2%E0%B9%89%E0%B8%AD%E0%B8%99%E0%B8%AB%E0%B8%A5%E0%B8%B1%E0%B8%87-10-%E0%B8%9B%E0%B8%B5.aspx?mode=year-range&value=30"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "lxml")
```

```python
data_list = soup.find(id="dl_lottery_stats_list")
```

```python
table = data_list.find_all('tr')
```

```python
df = pd.read_html(str(data_list))[0]

df_split = df[0].str.split(' ', expand=True)
drop_columns = [0,2,4,9,14,15,16,17,18,19]
filtered_df = df_split.drop(columns=drop_columns)

filtered_df.columns = ["month", "year", "รางวัลที่ 1", "2 ตัวบน", "3 ตัวบน", "2 ตัวล่าง", "3 ตัวหน้า 1", "3 ตัวหน้า 2", "3 ตัวล่าง 1", "3 ตัวล่าง 2"]

display(filtered_df)
```

# Save data to CSV file

```python
filtered_df.to_csv('./thai-lottery-statistics-30-years.csv')
```
